```python
import numpy as np
data = np.genfromtxt('indian_cricket_stats.csv', delimiter=',',skip_header=1, usecols=np.arange(1,89),dtype=str)
```

```python
runwick = np.empty([2000,2],dtype=int)

for i in range(0,300):
    if(data[i][54]=='-' or data[i][54]=='0' or data[i][54]==''):
        data[i][54]='-1'
    if(data[i][4]=='-' or data[i][3]=='0' or data[i][4]==''):
        data[i][4]='-1'
    if(data[i][2]=='-' or data[i][2]=='0' or data[i][2]==''):
        data[i][2]='-1'
    if(data[i][1]==''):
        data[i][1]=0
    runwick[i][0]=data[i][4]
    runwick[i][1]=data[i][54]

    temp = runwick[i][0]/runwick[i][1]  

    if(temp>=4 and temp<99 or temp>100 or temp<0):
        player_clasifier='Batsman'
    elif(temp<0.4 and temp>0.1):
        player_clasifier='All-rounder'
    elif(temp==1):
        player_clasifier='Not Capped'
    else:
        player_clasifier='Bowler'
    print(data[i][0]," ", player_clasifier)
    avg=(int)(data[i][4])/(int)(data[i][2]);
    print("Average : ", avg,"\n")
```
